```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.portalinmobiliario.com/MLC-618072768-amplia-casa-esquina-en-villa-el-vergel-maipu-_JM#position=1&search_layout=stack&type=item&tracking_id=aceba5bc-3b97-4ab3-85bc-1f839fddf8cc'
```

```python
propiedad_prueba = requests.get(url)
```

```python
propiedad_prueba.status_code
```

```python
s_prop_prueba = BeautifulSoup(propiedad_prueba.text, 'lxml')
```

```python
print(s_prop_prueba.prettify())
```

```python
tablas = s_prop_prueba.find('table', attrs={'class':'andes-table'}).find_all('td')#Buscamos (filtramos) según las secciones que nos interesen.

```

```python
print(tablas)
```

```python
tabla = tablas[0]
print(tabla)
```
